# Pymaceuticals Inc
As their Chief Data Analyst, you've been given access to the complete data from their most recent animal study. In this study, 250 mice were treated through a variety of drug regimes over the course of 45 days. Their physiological responses were then monitored over the course of that time. Your objective is to analyze the data to show how four treatments (Capomulin, Infubinol, Ketapril, and Placebo) compare.

To do this you are tasked with:


* Creating a scatter plot that shows how the tumor volume changes over time for each treatment.
* Creating a scatter plot that shows how the number of metastatic (cancer spreading) sites changes over time for each treatment.
* Creating a scatter plot that shows the number of mice still alive through the course of treatment (Survival Rate)
* Creating a bar graph that compares the total % tumor volume change for each drug across the full 45 days.

* Your legends should not be overlaid on top of any data.
* Your bar graph should indicate tumor growth as red and tumor reduction as green.
* It should also include a label with the percentage change for each bar. You may want to consult this tutorial for relevant code snippets.

As final considerations:

* You must include a written description of three observable trends based on the data.
* You must use proper labeling of your plots, including aspects like: Plot Titles, Axes Labels, Legend Labels, X and Y Axis Limits, etc.
* Your scatter plots must include error bars. This will allow the company to account for variability between mice. You may want to look into pandas.DataFrame.sem for ideas on how to calculate this.
* Remember when making your plots to consider aesthetics!

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import linregress
from scipy.stats import sem
```

```python
file1 = "clinicaltrial_data.csv"
file2 = "mouse_drug_data.csv"

clinical_data = pd.read_csv(file1)
mouse_data = pd.read_csv(file2)
```

```python
clinical_data.head()
```

```python
mouse_data.head()
```

```python
merge_table = pd.merge(clinical_data, mouse_data, on="Mouse ID")
merge_table
```

```python
drug_group = merge_table.groupby(["Drug", "Timepoint"])

capomulin_volume = drug_group["Tumor Volume (mm3)"].mean().loc["Capomulin"]
infubinol_volume = drug_group["Tumor Volume (mm3)"].mean().loc["Infubinol"]
ketapril_volume = drug_group["Tumor Volume (mm3)"].mean().loc["Ketapril"]
placebo_volume = drug_group["Tumor Volume (mm3)"].mean().loc["Placebo"]
```

```python
time = np.arange(0,50,5)

(c_slope, c_intercept, _, _, _) = linregress(time, capomulin_volume)
fit_capomulin = c_slope * time + c_intercept

(i_slope, i_intercept, _, _, _) = linregress(time, infubinol_volume)
fit_infubinol = i_slope * time + i_intercept

(k_slope, k_intercept, _, _, _) = linregress(time, ketapril_volume)
fit_ketapril = k_slope * time + k_intercept

(p_slope, p_intercept, _, _, _) = linregress(time, placebo_volume)
fit_placebo = p_slope * time + p_intercept
```

```python
#capomulin_new = capomulin_volume.reset_index(drop=True)
#blah = list(capomulin_new.values.flatten())
c_standard_errors = sem(capomulin_volume)
i_standard_errors = sem(infubinol_volume)
k_standard_errors = sem(ketapril_volume)
p_standard_errors = sem(placebo_volume)
```

```python
#Creating a scatter plot that shows how the tumor volume changes over time for each treatment.
#Capomulin, Infubinol, Ketapril, and Placebo


fig, ax = plt.subplots()

plt.scatter(time, capomulin_volume, marker="D", facecolors="turquoise", edgecolors="black", alpha=.75, label = "Capomulin") 
plt.scatter(time, infubinol_volume, marker="^", facecolors="red", edgecolors="black", alpha=.75, label = "Infubinol") 
plt.scatter(time, ketapril_volume, marker="s", facecolors="orange", edgecolors="black", alpha=.75, label = "Ketapril") 
plt.scatter(time, placebo_volume, marker="o", facecolors="green", edgecolors="black", alpha=.75, label = "Placebo")

ax.plot(time, fit_capomulin, 'b--', linewidth=.5)
ax.plot(time, fit_infubinol, 'r--', linewidth=.5)
ax.plot(time, fit_ketapril, 'y--', linewidth=.5)
ax.plot(time, fit_placebo, 'g--', linewidth=.5)

ax.errorbar(time, capomulin_volume, yerr = c_standard_errors, color="b", label="_", \
            linewidth=.5, fmt="none", capsize=5, alpha=.5)
ax.errorbar(time, infubinol_volume, yerr = i_standard_errors, color="r", label="_", \
            linewidth=.5, fmt="none", capsize=5, alpha=.5)
ax.errorbar(time, ketapril_volume, yerr = k_standard_errors, color="y", label="_", \
            linewidth=.5, fmt="none", capsize=5, alpha=.5)
ax.errorbar(time, placebo_volume, yerr = p_standard_errors, color="g", label="_", \
            linewidth=.5, fmt="none", capsize=5, alpha=.5)

fig.suptitle("Tumor Volume Over Time", fontsize = 16, fontweight="bold")
ax.set_xlabel("Time (Days)")
ax.set_ylabel("Tumor Volume (mm3)")


plt.legend()
plt.show()
```

```python
#Creating a scatter plot that shows how the number of metastatic (cancer spreading) sites changes over time for each treatment.
capomulin_metastatic = drug_group["Metastatic Sites"].mean().loc["Capomulin"]
infubinol_metastatic = drug_group["Metastatic Sites"].mean().loc["Infubinol"]
ketapril_metastatic = drug_group["Metastatic Sites"].mean().loc["Ketapril"]
placebo_metastatic = drug_group["Metastatic Sites"].mean().loc["Placebo"]

fig, ax = plt.subplots()

plt.plot(time, capomulin_metastatic, marker="D", c="turquoise",alpha=.75, label = "Capomulin") 
plt.plot(time, infubinol_metastatic, marker="^", c="red", alpha=.75, label = "Infubinol") 
plt.plot(time, ketapril_metastatic, marker="s", c="orange", alpha=.75, label = "Ketapril") 
plt.plot(time, placebo_metastatic, marker="o", c="green", alpha=.75, label = "Placebo")

#ax.plot(time, fit_capomulin, 'b--', linewidth=.5)
#ax.plot(time, fit_infubinol, 'r--', linewidth=.5)
#ax.plot(time, fit_ketapril, 'y--', linewidth=.5)
#ax.plot(time, fit_placebo, 'g--', linewidth=.5)

#ax.errorbar(time, capomulin_volume, yerr = c_standard_errors, color="b", label="_", \
#            linewidth=.5, fmt="none", capsize=5, alpha=.5)
#ax.errorbar(time, infubinol_volume, yerr = i_standard_errors, color="r", label="_", \
#            linewidth=.5, fmt="none", capsize=5, alpha=.5)
#ax.errorbar(time, ketapril_volume, yerr = k_standard_errors, color="y", label="_", \
#            linewidth=.5, fmt="none", capsize=5, alpha=.5)
#ax.errorbar(time, placebo_volume, yerr = p_standard_errors, color="g", label="_", \
#            linewidth=.5, fmt="none", capsize=5, alpha=.5)

fig.suptitle("Metastatic Sites Over Time", fontsize = 16, fontweight="bold")
ax.set_xlabel("Time (Days)")
ax.set_ylabel("Metastatic Sites")


plt.legend()
plt.show()
```
